```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/TP1/csv/futbol_argentino.csv")
```

```python
#Queremos la cantidad de goles en cada victoria de cada club grande.
#Violin plot

futbol = df[ ['local_team', 'local_result', 'visitor_result', 'visitor_team'] ]
clubes_grandes = ['Boca Juniors', 'Racing Club', 'River Plate', 'San Lorenzo', 'Independiente', 'Huracan']

futbol = futbol[ futbol['local_team'].isin(clubes_grandes) | futbol['visitor_team'].isin(clubes_grandes) ] #Filtramos por partidos donde haya clubes grandes.

futbol = futbol[ (futbol['local_team'].isin(clubes_grandes) & futbol['local_result'] > futbol['visitor_result']) | (futbol['visitor_team'].isin(clubes_grandes) & futbol['visitor_result'] > futbol['local_result']) ]
#Filtramos por partidos donde haya ganado el club grande.

ganadores_locales = futbol[ futbol['local_team'].isin(clubes_grandes) ][ ['local_team', 'local_result'] ].rename(columns={'local_team':'team', 'local_result':'result'}) #Filtramos por partidos ganados de local.
ganadores_visitantes = futbol[ futbol['visitor_team'].isin(clubes_grandes) ][ ['visitor_team', 'visitor_result'] ].rename(columns={'visitor_team':'team', 'visitor_result':'result'}) #Filtramos por partidos ganados de visitante.
ganadores = pd.concat([ganadores_locales, ganadores_visitantes]) #Juntamos goles de local y visitante del equipo ganador.
ganadores = ganadores[(ganadores != 0).all(1)] #Eliminamos ceros.

ganadores

#plt.figure(dpi=125)
#colores = ['#78C850', '#F08030',  '#6890F0',  '#F8D030', '#C03028', '#98D8D8']
#sns.violinplot(data=ganadores, x='team', y='result', palette=colores)
#plt.xticks(rotation=90)
#plt.xlabel("")
#plt.ylabel("Goles")
#plt.title("Distribución de cantidad de goles en victorias de equipos grandes")
```

```python
#Queremos los puntos por torneo de cada equipo grande.

futbol = df[ ['date_name', 'local_team', 'local_result', 'visitor_result', 'visitor_team'] ]
futbol['date_name'] = futbol['date_name'].apply(lambda x: x.split('-')[0].rstrip(' '))
clubes_grandes = ['Boca Juniors', 'Racing Club', 'River Plate', 'San Lorenzo', 'Independiente', 'Huracan']

futbol = futbol[ futbol['local_team'].isin(clubes_grandes) | futbol['visitor_team'].isin(clubes_grandes) ] #Filtramos por partidos donde haya clubes grandes.
victorias = futbol[ ((futbol['local_team'].isin(clubes_grandes)) & (futbol['local_result'] > futbol['visitor_result']))  | ((futbol['visitor_team'].isin(clubes_grandes)) & (futbol['visitor_result'] > futbol['local_result']))]

ganadores_locales = victorias[ victorias['local_team'].isin(clubes_grandes) ][ ['date_name', 'local_team', 'local_result'] ].rename(columns={'local_team':'team', 'local_result':'result'}) #Filtramos por partidos ganados de local.
ganadores_visitantes = victorias[ victorias['visitor_team'].isin(clubes_grandes) ][ ['date_name', 'visitor_team', 'visitor_result'] ].rename(columns={'visitor_team':'team', 'visitor_result':'result'}) #Filtramos por partidos ganados de visitante.
ganadores = pd.concat([ganadores_locales, ganadores_visitantes]) #Juntamos goles de local y visitante del equipo ganador.
ganadores = ganadores[(ganadores != 0).all(1)] #Eliminamos ceros.

res = ganadores.groupby(['date_name', 'team']).size().to_frame().reset_index().rename(columns={0:'triunfos'})

#plt.figure(dpi=125)
#colores = ['#78C850', '#F08030',  '#6890F0',  '#F8D030', '#C03028', '#98D8D8']
#sns.boxplot(data=res, x='team', y='triunfos', palette=colores)
#plt.title("Distribución de triunfos de equipos grandes en todos los torneos")
#plt.xticks(rotation=90)
#plt.xlabel("Equipo")
#plt.ylabel("Triunfos")

res
```
